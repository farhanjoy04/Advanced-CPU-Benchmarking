```python
import platform
import psutil
import time
import matplotlib.pyplot as plt
import numpy as np

def fetch_cpu_info():
    """Fetch CPU information including processor name and architecture."""
    system_info = {
        "Processor Name": platform.processor(),
        "System": platform.system(),
        "Version": platform.version(),
        "Architecture": platform.architecture()[0],
        "CPU Cores": psutil.cpu_count(logical=False),
        "CPU Threads": psutil.cpu_count(logical=True),
        "Logical processors": psutil.cpu_count(logical=True),
        "Max Frequency": psutil.cpu_freq().max,
    }
    return system_info

def benchmark_cpu():
    """Benchmark CPU by calculating the CPU utilization percentage."""
    def is_prime(n):
        if n < 2:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True
    start_time = time.time()
    primes = [x for x in range(1, 100000) if is_prime(x)]
    end_time = time.time()
    execution_time = end_time - start_time
    score = int(100000 / execution_time)
    return score, execution_time

def benchmark_memory(data_size_mb=100):
    """Benchmark memory bandwidth with data size option."""
    data_size = data_size_mb * 1024 * 1024
    data = b'a' * data_size

    # Write test
    start_time = time.time()
    memoryview(data)  # Force data into memory
    end_time = time.time()
    write_bandwidth = data_size / (end_time - start_time)

    # Read test
    start_time = time.time()
    _ = sum(data)  # Access data elements to simulate reading
    end_time = time.time()
    read_bandwidth = data_size / (end_time - start_time)

    return write_bandwidth, read_bandwidth

def visualize_benchmark(cpu_results, memory_results):
    """Generate visual representations of benchmarking results."""
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # CPU Benchmark Bar Chart
    cpu_labels = list(cpu_results.keys())
    cpu_values = list(cpu_results.values())
    axes[0].bar(cpu_labels, cpu_values, color="skyblue")
    axes[0].set_title("CPU Benchmarking Score", fontsize=14)
    axes[0].set_ylabel("Score", fontsize=12)
    axes[0].set_xlabel("Metrics", fontsize=12)
    axes[0].grid(axis="y", linestyle="--", alpha=0.7)

    # Memory Benchmark Bar Chart
    memory_labels = ['Write Bandwidth (MB/s)', 'Read Bandwidth (MB/s)']
    memory_values = [memory_results[0] / (1024 * 1024),
                     memory_results[1] / (1024 * 1024)]
    axes[1].bar(memory_labels, memory_values, color="lightgreen")
    axes[1].set_title("Memory Bandwidth", fontsize=14)
    axes[1].set_ylabel("Bandwidth (MB/s)", fontsize=12)
    axes[1].grid(axis="y", linestyle="--", alpha=0.7)

    plt.tight_layout()
    plt.show()


def visualize_performance():
    """Simulate and plot CPU usage over time."""
    usage_data = []
    timestamps = []
    for _ in range(10):
        usage = psutil.cpu_percent(interval=1)
        usage_data.append(usage)
        timestamps.append(time.time())
    readable_times = [time.strftime("%H:%M:%S", time.localtime(ts)) for ts in timestamps]
    plt.figure(figsize=(10, 6))
    plt.plot(readable_times, usage_data, marker="o", linestyle="-", color="green")
    plt.title("CPU Usage Over Time", fontsize=16)
    plt.ylabel("CPU Usage (%)", fontsize=12)
    plt.xlabel("Time", fontsize=12)
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def visualize_memory_benchmark(data_sizes, write_bandwidths, read_bandwidths):
    """Generates a bar chart to visualize memory benchmark results."""
    width = 0.35  # the width of the bars
    x = np.arange(len(data_sizes))  # the label locations

    fig, ax = plt.subplots(figsize=(10, 6))
    rects1 = ax.bar(x - width/2, write_bandwidths, width, label='Write Bandwidth (MB/s)', color='skyblue')
    rects2 = ax.bar(x + width/2, read_bandwidths, width, label='Read Bandwidth (MB/s)', color='lightgreen')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Bandwidth (MB/s)')
    ax.set_title('Memory Bandwidth Benchmark')
    ax.set_xticks(x)
    ax.set_xticklabels(data_sizes)
    ax.legend()

    # Add value labels to the bars
    def auto_text(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(round(height, 2)),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    auto_text(rects1)
    auto_text(rects2)

    fig.tight_layout()
    plt.show()

def main():
    print("Fetching CPU Information...")
    cpu_info = fetch_cpu_info()
    for key, value in cpu_info.items():
        print(f"{key}: {value}")

    print("\nStarting CPU Benchmark...")
    benchmark_score, execution_time = benchmark_cpu()
    print(f"Benchmark Score: {benchmark_score}")
    print(f"Execution Time: {execution_time:.2f} seconds")

    data_sizes = [10, 100, 1000]  # Data sizes in MB
    write_bandwidths = []
    read_bandwidths = []
    for size in data_sizes:
        print(f"\nStarting Memory Benchmark (Data Size: {size}MB)...")
        write_bandwidth, read_bandwidth = benchmark_memory(size)
        write_bandwidths.append(write_bandwidth / (1024 * 1024))  # Convert to MB/s
        read_bandwidths.append(read_bandwidth / (1024 * 1024))  # Convert to MB/s
        print(f"Write Bandwidth: {write_bandwidth / (1024 * 1024):.2f} MB/s")
        print(f"Read Bandwidth: {read_bandwidth / (1024 * 1024):.2f} MB/s")

        # Prepare results for visualization
        memory_results = (write_bandwidth, read_bandwidth)
        benchmark_results = {
            "Benchmark Score": benchmark_score,
            "Execution Time (s)": execution_time,
        }

        print("\nVisualizing Benchmark Results...")
        visualize_benchmark(benchmark_results, memory_results)

    # Generate the memory benchmark graph
    visualize_memory_benchmark(data_sizes, write_bandwidths, read_bandwidths)

    print("\nVisualizing CPU Performance...")
    visualize_performance()

if __name__ == "__main__":
    main()
```
